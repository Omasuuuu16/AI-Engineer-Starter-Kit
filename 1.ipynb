```python
url = "https://official-joke-api.appspot.com/random_joke"

response = requests.get(url)

print(response)
```

```python
print(response.status_code)
```

Status Codes
Code	المعنى

200	    نجح

201 	تم إنشاء مورد جديد  

400	    الطلب فيه خطأ

401 	محتاج تسجيل دخول أو API Key

403 	ممنوع

404	    الصفحة غير موجودة

500 	مشكلة في السيرفر

HuggingFace

```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis") #تحليل المشاعر
```

```python
classifier("I love AI")
```

```python
classifier("I want to cry")
```

```python
sentence=["I love myself","I am happy"]
classifier(sentence)
```

أهم مهام pipeline()

المهمة	الكود

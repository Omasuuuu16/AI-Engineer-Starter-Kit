```python
import requests
```

```python
from transformers import pipeline
```

```python
response=requests.get("https://official-joke-api.appspot.com/random_joke")

data = response.json()
data
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
txt= data["setup"]+ " " + data["punchline"]
classifier(txt)
```
